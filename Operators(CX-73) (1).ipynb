```python
# %= :-- Modulus Equal to

x = 20
y = 30
y %= x  #(y = reminder)y/x
z = x+(y*10)
print(y)
print(z)
```

```python
# **= :-- Exponent Equal to

x = 20
y = 30
y **= x  # y = y^x
z = x+(y*10)
print(y)
print(z)
```

```python
# //= :-- Floor Equal to

x = 20
y = 10
y //= x  #(y = floor y/x)
z = x+(y*10)
print(y)
print(z)


```

```python
#Logical Operators :--
# Logical AND
# logical OR
# logical NOT


# Logical AND :-- If both the operands(Conditions) are true the condition becomes True otherwise it is false


# True = 1
# False = 0

# A      B      O/P
# True   True    True
# True   False   False
# False  True    False
# False  False   False


#Logical OR :-- If any one of the Operands(Conditions) is true then it is True otherwise it is False
# A      B      O/P
# True   True    True
# True   False   True
# False  True    True
# False  False   False


# Logical NOT :-- It is used to reverse the lofical state of its Operand
# A              O/P
# True           False
# False          True

```

```python
first_con = 20
second_con = 10

print(first_con and second_con)


first_con = 20
second_con = 10

print(first_con or second_con)


first_con = True
second_con = False

print(not first_con ,  not second_con)
```
