```python
def swap_and_concat_strings(str1, str2):
    # Swap the first two characters of each string
    swapped_str1 = str2[:2] + str1[2:]
    swapped_str2 = str1[:2] + str2[2:]

    # Concatenate the swapped strings with 'a' in between
    result = swapped_str1 + 'a' + swapped_str2
    return result

# Example usage:
string1 = 'abc'
string2 = 'xyz'
output = swap_and_concat_strings(string1, string2)
print(output)
```
